   ### Q1: What are the best and worst performing cuisine categories in New york?

```python
df2.head(10)
```

```python
with client:
    db = client.Scenario3
    qresult = db.restaurants.aggregate([
        {"$match": {"borough": {"$in": ["Brooklyn", "Manhattan", "Staten Island", "Bronx", "Queens"]}, "cuisine": {"$ne": None}}},
        {"$group": {
            "_id": "$cuisine",
            "avg_score": {"$avg": "$grades.score"},
            "recent_score": {"$first": "$grades.score"},
            "recent_grade": {"$first": "$grades.grade"}
        }},
        {"$project": {
            "cuisine": "$_id",
            "avg_score": {"$ifNull": ["$avg_score", 0]},
            "recent_score": 1,
            "recent_grade": 1,
            "_id": 0
        }},
        {"$sort": {"avg_score": -1}}
    ])

    for result in qresult:
        cuisine = result['cuisine']
        avg_score = result['avg_score']
        recent_score = result['recent_score']
        recent_grade = result['recent_grade']
        
        print(f"Cuisine: {cuisine}")
        print(f"Average Score: {avg_score}")
        print(f"Recent Score: {recent_score}")
        print(f"Recent Grade: {recent_grade}")
        print("------------------------")
```
